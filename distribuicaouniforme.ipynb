#Distribuição Uniforme Contínua

 **Definição:**  Uma variável aleatória tem distribuição Uniforme Contínua se atender
as condições abaixo:
1. O resultado é um número real em um intervalo de
limites conhecidos a e b;
2. Todos os valores dentro do intervalo têm igual
probabilidade de ocorrência.

**representação:**  X ~ U [a , b]

```python
import pandas as pd
import math
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Definindo os parâmetros da distribuição uniforme
a = 0   # Início do intervalo
b = 10  # Fim do intervalo
t = 0.0012 # parâmetro da fgm
```

#Função Geradora de Momentos


É representada pela fórmula:


fgm - du.png

```python
#função geradora de momentos
a=0
b=10
t=0.0012
fgm = (sp.exp(t*b) - sp.exp(t*a))/t*(b-a)
print(f'A função geradora de momentos é:',fgm)
```

#Esperança

A esperança da distribuição é representado pela fórmula:


esp - du.png

```python
esp= (a+b)/2
print(f'A esperança é:', esp)
```

#Esperança ao quadrado

 A esperança da distribuição é representado por:


 esp2 - du.png

```python
esp_quad= (a**3 - b**3)/(3*(a-b))
print(f'A esperança ao quadrado é:', esp_quad)
```

#Variância

É representado matematicamente por:


var - du.png

```python
var= ((a-b)**2)/12
print(f'A variância é:', var)
```

```python
#gerando numeros aleatorios
import random
aleat= []
for i in range (1000):
  x= random.uniform (2,4)
  aleat.append(x)
print(x)
```

```python
#histogrrama da distribuição uniforme
plt.subplot(211)
plt.plot(aleat)
plt.xlabel('eixo horizontal')
plt.ylabel('números aleátorios entre 2 e 4')

plt.subplot(212)
plt.hist(aleat, bins=15)
plt.xlabel('Classe')
plt.ylabel('frequência')
```
